```python
#import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as lines
import matplotlib.ticker as tick
from matplotlib.ticker import FormatStrFormatter
from matplotlib.legend_handler import HandlerLineCollection, HandlerTuple

from matplotlib.pyplot import rc
rc("text", usetex=True)
rc("font", **{"family":"sans-serif","sans-serif":["Helvetica"]},size="14")
rc("text.latex",preamble=r"\usepackage{sfmath}")

%matplotlib notebook
```

```python
# import data
# change number of variables according to current dataset
# change/delete delimiter and skiprows as necessary

#PDF data + refinement
r, G_trunc, G_diff, G_calc = \
        np.loadtxt("C:/Users/sincl/Documents/MaughanLab/Kobi/PDF_refinements/Li3InCl6_cell3_no_mixing_refinement.txt",
        unpack=True, dtype=float, delimiter=" ", skiprows=4)
```

```python
# generates plot
plt.figure(figsize=(7,7))

# plot data
plt.plot(r, G_trunc, color="black", label="Observed", marker=".")
plt.plot(r, G_calc, color="#FA8072", label="Calculated", linewidth="2")
plt.plot(r, G_diff-1, color="#BEBEBE", label="Difference")

# set axis limits
plt.xlim(1, 22)
plt.ylim(-1.25, 2)

# set axis labels
plt.xlabel("r / " r"$\AA$")
plt.ylabel("G(r) / " r"$\AA^{-2}$")

# set title
plt.text(1.2, 1.86, r"Li$_3$InCl$_6$, No site mixing")

# add text for R-value
plt.text(17.6, 1.35, "R" r"$_W=$ 11.45\%")

plt.legend()
plt.tight_layout()

#plt.savefig("C:/Users/sincl/Desktop/Li3InCl6_cell3_no_mixing.png",bbox_inches='tight',dpi=1000) 
```
